```python
import pandas as pd

df_lone= pd.read_csv("DATA/MSOA-Lone Parent Households.csv")
df_lone.head()
```

```python
# rename
df_lone = df_lone.rename(columns={
    'Middle layer Super Output Areas Code': 'MSOA',
    'Household composition (15 categories)': 'composition',
    'Observation': 'count'
})

# change to numeric
df_lone['count'] = pd.to_numeric(df_lone['count'], errors='coerce')

# clean invalid 
df_valid = df_lone[df_lone['composition'] != 'Does not apply'].copy()

# add lone parent 
df_valid.loc[:, 'is_lone_parent'] = df_valid['composition'].str.contains("Lone parent", case=False, na=False)

# count total count and lone count
total = df_valid.groupby('MSOA')['count'].sum()
lone = df_valid[df_valid['is_lone_parent']].groupby('MSOA')['count'].sum()

# calculate
df_lone_clean = pd.DataFrame({
    'lone_parent_percent': (lone / total) * 100
}).reset_index()
```

```python
# output
df_lone_clean.head()
df_lone_clean.to_csv("cleaned/cleaned_lone.csv", index=False)
```
