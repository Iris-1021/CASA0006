```python
import pandas as pd

df_commute = pd.read_csv("DATA/MSOA-distance travelled to work.csv")
df_commute.head()
```

```python
# rename and change to numeic
df_commute = df_commute.rename(columns={
    'Middle layer Super Output Areas Code': 'MSOA',
    'Distance travelled to work (5 categories)': 'distance',
    'Observation': 'count'
})

df_commute['count'] = pd.to_numeric(df_commute['count'], errors='coerce')

# define the long distance
long_distances = ['10km to less than 30km', '30km and over']

# excluded commuting 
excluded_distances = ['Works mainly from home',
                      'Not in employment or works mainly offshore, in no fixed place or outside the UK']

# fliter commuting 
df_commuting = df_commute[~df_commute['distance'].isin(excluded_distances)].copy()

# count the number of commuting in MSOA 
total_commuters = df_commuting.groupby('MSOA')['count'].sum()

# count the number of long distance commuting
long_commuters = df_commuting[df_commuting['distance'].isin(long_distances)].groupby('MSOA')['count'].sum()

# calculate 
df_commute_clean = pd.DataFrame({
    'long_commute_percent': (long_commuters / total_commuters) * 100
}).reset_index()
```

```python
# output
df_commute_clean.head()
df_commute_clean.to_csv("cleaned/commute_cleaned.csv", index=False)
```
