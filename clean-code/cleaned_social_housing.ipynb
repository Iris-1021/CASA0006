```python
import pandas as pd
df_social = pd.read_csv("DATA/MSOA-Social Housing.csv")
df_social.head()
```

```python
# rename
df_social = df_social.rename(columns={
    'Middle layer Super Output Areas Code': 'MSOA',
    'Tenure of household (9 categories)': 'tenure',
    'Observation': 'count'
})

# change to numeric
df_social['count'] = pd.to_numeric(df_social['count'], errors='coerce')

# delete invalid
df_valid = df_social[df_social['tenure'] != 'Does not apply'].copy()

# define social house
df_valid['is_social'] = df_valid['tenure'].str.contains("Social rented", case=False, na=False)

# count total and social
total = df_valid.groupby('MSOA')['count'].sum()
social = df_valid[df_valid['is_social']].groupby('MSOA')['count'].sum()

# calculate percent of social housing
df_social_clean = pd.DataFrame({
    'social_housing_percent': (social / total) * 100
}).reset_index()
```

```python
# output
df_social_clean.head()
df_social_clean.to_csv("cleaned/social_housing_cleaned.csv", index=False)
```
